```python
import json
import pandas as pd
from google.colab import files


# Opeing the json file from week 10
# https://github.com/DataScienceLiam/DataScienceLiam.github.io/blob/main/Gov%20spending%2C%20Education%20and%20Life%20expectancy.json
with open('/content/HDI, Gov spending, Education and Life expectancy.json') as f:
    data = json.load(f)
df_merged = pd.DataFrame(data)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Setting the X and Y values from the dataframe
X = df_merged[['Log Tax per Capita']]
y = df_merged[['HDI']]
```

```python
regressor = LinearRegression()
regressor.fit(X, y)


y_pred = regressor.predict(X)
# Calculates R^2
score = regressor.score(X,y)

```

```python
# Ploting the graph, choosing colours, adding a title, naming axis and displaying R^2
plt.annotate(f'R\u00b2 = {score:.2f}', xy=(0.1, 0.8), xycoords='axes fraction')
plt.scatter(X, y, color= '#CBC3E3')
plt.plot(X, regressor.predict(X), color='#710193')
plt.title('Log of Tax per capita vs HDI')
plt.xlabel('Log of Tax per capita')
plt.ylabel('HDI')
plt.show()
```
