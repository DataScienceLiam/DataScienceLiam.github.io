```python
import pandas as pd
```

```python
Wages_cost = pd.read_html('https://www.spotrac.com/epl/payroll/')
Wages_cost[0]
```

```python
df = Wages_cost[0]
df.head(5)
```

```python
df_simplified = df.drop(['Rank', 'Active Players', 'Forwards',
                        'Midfielders', 'Defensemen', 'Goalkeepers', 'Transfer Fees'], axis=1)
df_simplified.rename(columns = {'Est. Total Salary':'Estimated Total Salary'}, inplace = True)
```

```python
df_top16 = df_simplified.iloc[:16]
df_top16['Estimated Total Salary'] = df_top16['Estimated Total Salary'].str.replace(',', '').str.replace('£', '')

print(df_top16)
```

```python
exported_values = list(df_top16.T.to_dict().values())
```

```python
import json
open('Total_wage_cost_football_data.json' , 'w').write( json.dumps(exported_values) )
```
