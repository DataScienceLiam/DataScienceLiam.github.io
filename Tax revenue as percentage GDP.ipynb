```python
# This was a test to see the correlation between tax revenue and growth for certain nations
# Not in use


import pandas as pd
from pandas import DataFrame
from typing import Set, Any


def remove_others(df: DataFrame, columns: Set[Any]):
    cols_total: Set[Any] = set(df.columns)
    diff: Set[Any] = cols_total - columns
    df.drop(diff, axis=1, inplace=True)


df_GDP = pd.read_csv('gdp-per-capita-growth.csv')
df_GDP
```

```python
df_Tax = pd.read_csv('total-tax-revenues-gdp.csv')
df_Tax
```

```python
# df_simplified = df.drop(['Rank', 'Active Players', 'Forwards',
#                         'Midfielders', 'Defensemen', 'Goalkeepers', 'Transfer Fees'], axis=1)
# df_simplified.rename(columns = {'Est. Total Salary':'Estimated Total Salary'}, inplace = True)
```

```python
from functools import reduce
dfs = [df_GDP, df_Tax]
merged_df = reduce(lambda left, right: pd.merge(
    left, right, on=['Entity', 'Code', 'Year'], how='outer'), dfs)
```

```python
merged_df
```

```python
df_clean = merged_df.dropna()
df_clean = df_clean.reset_index(drop=True)
df_clean = df_clean[df_clean['GDP per capita growth (annual %)'] != 0.00]
df_clean = df_clean[df_clean['Total tax revenue (% of GDP) (ICTD (2021))'] != 0.00]
df_clean
```

```python
df = df_clean.drop_duplicates(subset=['Entity'], keep='last')
```

```python
df = df.reset_index(drop=True)
df
```

```python
df_new = df.loc[df['Code'].isin(['AUS', 'CAN','SAU', 'USA', 'IND', 'RUS', 'ZAF',  'TUR', 'ARG', 'BRA', 'MEX','FRA', 'DEU', 'ITA', 'GBR', 'CHN', 'IDN', 'JPN', 'KOR'])]

df_final = df_new.reset_index(drop=True)

df_final
```

```python
import json
exported_values = list(df_final.T.to_dict().values())
open('Tax vs GDP growth Data.json', 'w').write(json.dumps(exported_values))
```
